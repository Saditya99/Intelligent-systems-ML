```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"C:\Users\Sadit\Downloads\Ames_Housing_Sales.csv")

df.head()
```

```python
df.describe()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

correlations = numeric_df.corr()['SalePrice']

selected_features = correlations[abs(correlations) > 0.2].index.tolist()

data=df[selected_features]
data.head()
```

```python
cat=df.select_dtypes(include='string')
print(cat)
```

```python
print(df['BldgType'].unique())
```

```python
alley_dummies = pd.get_dummies(df['Alley'], prefix='A')
df = pd.concat([df, alley_dummies], axis=1)

data['BldgType'] = df['BldgType'].map({
    '1Fam': 1,
    '2fmCon': 2,
    'Duplex': 3,
    'TwnhsE': 4,
    'Twnhs': 5
})
data['LotShape'] = df['LotShape'].map({
    'Reg': 4,
    'IR1': 3,
    'IR2': 2,
    'IR3': 1
})
data['LotConfig'] = df['LotConfig'].map({
    'Inside': 1,
    'Corner': 2,
    'CulDSac': 3,
    'FR2': 4,
    'FR3': 5
})
data['LandContour'] = df['LandContour'].map({
    'Lvl': 3,
    'Bnk': 2,
    'HLS': 1,
    'Low': 0
})
data['LandSlope'] = df['LandSlope'].map({
    'Gtl': 3,
    'Mod': 2,
    'Sev': 1
})

roofstyle_dummies = pd.get_dummies(df['RoofStyle'], prefix='RoofStyle')
data = pd.concat([data, roofstyle_dummies], axis=1)

data['ExterQual'] = df['ExterQual'].map({
    'Ex': 4,
    'Gd': 3,
    'TA': 2,
    'Fa': 1,
    'Po': 0
})
data['ExterCond'] = df['ExterCond'].map({
    'Ex': 4,
    'Gd': 3,
    'TA': 2,
    'Fa': 1,
    'Po': 0
})
data['BsmtQual'] = df['BsmtQual'].map({
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1
}).fillna(0)
data['BsmtCond'] = df['BsmtCond'].map({
    'Ex': 4,
    'Gd': 3,
    'TA': 2,
    'Fa': 1,
    'Po': 0
}).fillna(0)

data.head()
```

```python
finalcor = data.corr()['SalePrice'].drop('SalePrice')
finalcor = finalcor[abs(finalcor) > 0.5]

selected_columns = finalcor.index.tolist()

data = data[selected_columns]
data['SalePrice']=df['SalePrice']
data.head()
```

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)

IQR=Q3-Q1

lower=Q1-IQR*6.0
upper=Q1+IQR*6.0

data=data[(data>lower) & (data<upper)]
print(len(df))
```

```python
data['SalePrice'] = data['SalePrice'].fillna(data['SalePrice'].mean())
data = data.fillna(data.mode().iloc[0])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler

X = data.drop(['SalePrice'], axis=1)
scaler = StandardScaler()
X = scaler.fit_transform(X)

Y = data['SalePrice']

score = []

limit = int(len(data) ** 0.5)

for i in range(1, limit,2):
    model = KNeighborsRegressor(n_neighbors=i,weights='distance')
    eval_scores = cross_val_score(model, X, Y, cv=5)
    score.append([i, eval_scores.mean()])

bestK = max(score, key=lambda x: x[1])[0]
bestS=max(score,key=lambda x :x[1])[1]
print(bestK)
print(bestS)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=69)


finalModel = KNeighborsRegressor(n_neighbors=bestK,weights='distance')
finalModel.fit(X_train,Y_train)


print(f"Model: KNN with k={bestK}")
print(f"Training R²: {finalModel.score(X_test, Y_test):.4f}")
```

```python
import joblib
joblib.dump(finalModel, 'house_price_knn.pkl')
```
